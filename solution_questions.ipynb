## 1. Load the Data

```python
import pandas as pd

data = pd.read_csv('Sweden_population_distribution.csv')

print(data.head())
```

## 2: Inspect the Data

```python
print(data.info())

print(data.isnull().sum())

print(data.describe())
```

## 1. What is the trend in Swedenâ€™s total population over time?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population'], marker='o', label='Total Population', color='blue')
plt.title('Total Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()
```

## 2. How do male and female populations compare annually?

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(len(data['Year']))
width = 0.4

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, data['Men'], width, label='Men', color='skyblue')
plt.bar(x + width/2, data['Women'], width, label='Women', color='pink')
plt.title('Annual Comparison of Male and Female Populations')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(x, data['Year'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. What are the trends in the age distribution categories (0-17, 18-65, 65+)?


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.stackplot(
    data['Year'], 
    data['0-17'], data['18-65'], data['65+'], 
    labels=['0-17', '18-65', '65+'], 
    alpha=0.8
)
plt.title('Age Distribution Categories Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper left')
plt.show()
```

## 4. How has the number of foreign-born residents changed over time?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Foreign-born'], marker='o', label='Foreign-born', color='orange')
plt.title('Foreign-Born Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Foreign-Born Population')
plt.grid(True)
plt.legend()
plt.show()
```

## 5. What is the trend in life expectancy for men and women?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Life expectancy, men'], marker='o', label='Men', color='green')
plt.plot(data['Year'], data['Life expectancy, women'], marker='o', label='Women', color='purple')
plt.title('Life Expectancy for Men and Women Over the Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend()
plt.grid(True)
plt.show()
```

## 6. How do birth and death rates affect population growth?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(data['Year'], data['Births'], marker='o', label='Births', color='blue')
plt.plot(data['Year'], data['Deaths'], marker='o', label='Deaths', color='red')
plt.title('Births vs Deaths Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

## 7. How has the total fertility rate fluctuated over the years?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Total fertility rate'], marker='o', label='Total Fertility Rate', color='cyan')
plt.title('Total Fertility Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Fertility Rate')
plt.grid(True)
plt.legend()
plt.show()
```

## 8. What is the net migration trend (in-migration vs. out-migration)?


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(data['Year'], data['In-migration'], label='In-migration', alpha=0.7, color='green')
plt.bar(data['Year'], -data['Out-migration'], label='Out-migration', alpha=0.7, color='red')
plt.title('In-Migration vs Out-Migration Annually')
plt.xlabel('Year')
plt.ylabel('Population Change')
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()
```

 ## 9. Are there correlations between household size and population growth?

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(data['Household units'], data['Population growth'], color='purple', alpha=0.7)
plt.title('Household Size vs Population Growth')
plt.xlabel('Household Units')
plt.ylabel('Population Growth')
plt.grid(True)
plt.show()
```

## 10. What is the historical trend in marriage and divorce rates?

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Year'], data['Marriages'], width=0.4, label='Marriages', color='blue', align='center')
plt.bar(data['Year'], data['Divorces'], width=0.4, label='Divorces', color='orange', align='edge')
plt.title('Marriage and Divorce Rates by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```
