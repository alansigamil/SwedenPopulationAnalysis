## 1. Line Chart: Total Population Over the Years

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
data = pd.read_csv('Sweden_population_distribution.csv')

plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Population'], marker='o', color='blue', label='Total Population')
plt.title('Total Population Over the Years')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()
```

## 2. Bar Graph: Annual Comparison of Male and Female Populations

```python
import numpy as np

x = np.arange(len(data['Year']))
width = 0.4

plt.figure(figsize=(12, 6))
plt.bar(x - width/2, data['Men'], width, label='Men', color='skyblue')
plt.bar(x + width/2, data['Women'], width, label='Women', color='pink')
plt.title('Annual Comparison of Male and Female Populations')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(x, data['Year'], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

## 3. Stacked Area Chart: Age Distribution Categories Over Time

```python
plt.figure(figsize=(12, 6))
plt.stackplot(data['Year'], data['0-17'], data['18-65'], data['65+'], labels=['0-17', '18-65', '65+'], alpha=0.8)
plt.title('Age Distribution Categories Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper left')
plt.show()
```

## 4. Line Graph: Trends in Foreign-Born and Swedish Citizens Annually

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Foreign-born'], marker='o', label='Foreign-born', color='orange')
plt.plot(data['Year'], data['Swedish citizens'], marker='o', label='Swedish Citizens', color='green')
plt.title('Trends in Foreign-Born and Swedish Citizens Annually')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Dual Axis Line Graph: Birth Rates vs Death Rates Over Time

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.set_xlabel('Year')
ax1.set_ylabel('Births', color='blue')
ax1.plot(data['Year'], data['Births'], marker='o', color='blue', label='Births')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.set_ylabel('Deaths', color='red')
ax2.plot(data['Year'], data['Deaths'], marker='o', color='red', label='Deaths')
ax2.tick_params(axis='y', labelcolor='red')

plt.title('Birth Rates vs Death Rates Over Time')
fig.tight_layout()
plt.grid(True)
plt.show()
```

## 6. Scatter Plot: Household Size vs Population Growth

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Household units'], data['Population growth'], color='purple', alpha=0.7)
plt.title('Household Size vs Population Growth')
plt.xlabel('Household Units')
plt.ylabel('Population Growth')
plt.grid(True)
plt.show()
```

## 7. Bar Graph: Marriage and Divorce Rates by Year

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Year'], data['Marriages'], width=0.4, label='Marriages', color='blue', align='center')
plt.bar(data['Year'], data['Divorces'], width=0.4, label='Divorces', color='orange', align='edge')
plt.title('Marriage and Divorce Rates by Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.tight_layout()
plt.show()
```

## 8. Line Graph: Life Expectancy for Men and Women Over the Years

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Year'], data['Life expectancy, men'], marker='o', label='Men', color='green')
plt.plot(data['Year'], data['Life expectancy, women'], marker='o', label='Women', color='purple')
plt.title('Life Expectancy for Men and Women Over the Years')
plt.xlabel('Year')
plt.ylabel('Life Expectancy (Years)')
plt.legend()
plt.grid(True)
plt.show()
```

## 9. Bar Graph: In-Migration vs Out-Migration Annually

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Year'], data['In-migration'], label='In-migration', alpha=0.7, color='green')
plt.bar(data['Year'], -data['Out-migration'], label='Out-migration', alpha=0.7, color='red')
plt.title('In-Migration vs Out-Migration Annually')
plt.xlabel('Year')
plt.ylabel('Population Change')
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.show()
```

## 10. Heatmap: Correlation Matrix for Numerical Variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv('Sweden_population_distribution.csv')

correlation_matrix = data.corr()

mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(14, 10))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', mask=mask,
            cbar_kws={'shrink': 0.8}, linewidths=0.8, vmin=-1, vmax=1,
            annot_kws={'size': 14, 'weight': 'bold', 'color': 'black'},  
            linecolor='white', cbar=True)  
plt.title('Correlation Matrix for Numerical Variables', fontsize=18, weight='bold')
plt.show()
```
